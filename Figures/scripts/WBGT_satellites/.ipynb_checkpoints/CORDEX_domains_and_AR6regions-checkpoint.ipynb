## Get the data

```python
from climaf.api import *
from IPython.display import Image
```

```python
pattern = '/data/jservon/IPCC/WBGT/CORDEX/${CORDEX_domain}/${member}_${thres}_${experiment}_${clim_period}.nc'
cproject('wbgt_cordex_ch12_test','experiment',('period','fx'),'clim_period','CORDEX_domain','member','thres', ('variable','WBGTindoor'), ensemble=['member'], separator='%')
dataloc(project='wbgt_cordex_ch12_test', url=pattern) 
```

```python
def return_CORDEX_file(CORDEX_domain):
    thres = 'Exceed33'
    req_dict = dict(project='wbgt_cordex_ch12_test',
                     experiment = 'rcp85',
                     clim_period = '3.0K',
                     CORDEX_domain = CORDEX_domain,
                     thres = thres
                    )
    req_exp = ds(member = '*',**req_dict)
    first_model = req_exp.explore('choices')['member'][0]
    test_dat = clim_average( ds(member = first_model, **req_dict), 'ANM')
    return regridn(test_dat, cdogrid='r360x180')
```

# AFR

```python
CORDEX_domain = 'AFR-22'
cordex_dat = return_CORDEX_file(CORDEX_domain)

trim_figure = '/home/jservon/Chapter12_IPCC/figs/CORDEX_domains_vs_AR6_regions/'+CORDEX_domain+'.png'

cmd = 'python /home/jservon/Chapter12_IPCC/scripts/WBGT_satellites/plot_CORDEX_domain_vs_AR6_regions.py '+CORDEX_domain+' '+cfile(cordex_dat)+' '+trim_figure
os.system(cmd)

Image(trim_figure)
```
