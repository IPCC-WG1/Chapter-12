## Work on the individual models:
- create scenario and baseline ensembles => select common models
- compute annual sums
- compute climatologies
- compute differences

```python
wthres = '33'
thres = 'Exceed'+wthres
variable = 'wbgt'

exp_list = [
    dict(experiment='ssp585',
         period = '2041-2060'),
    dict(experiment='ssp585',
         period = '2081-2100'),
    dict(experiment='ssp126',
         period = '2081-2100'),
]

# -- Create ensemble for historical baseline
req_baseline = ds(project='wbgt_cmip6_ch12',
                  experiment = 'ssp585',
                  period = '1995-2014',
                  member = '*',
                  thres = thres
                 )
ens_baseline = req_baseline.explore('ensemble')
#req_baseline.explore('choices')

# -- Loop on the scenarios
ens_diff = dict()
for exp in exp_list:
    #
    # -- Experiment and period
    experiment = exp['experiment']
    period = exp['period']
    
    # -- Create ensemble object for the scenario
    req_exp = ds(project='wbgt_cmip6_ch12',
                 experiment = experiment,
                 period = period,
                 member = '*',
                 thres = thres
                )
    ens_exp = req_exp.explore('ensemble')
    
    # -- Extract common members
    wens_baseline, wens_exp = ensemble_intersection([ens_baseline, ens_exp])
    #    
    # -- Climatologies
    clim_baseline = clim_average(wens_baseline, 'ANM')
    clim_exp      = clim_average(wens_exp, 'ANM')
    
    # -- Changes = Scenario minus baselines
    diff_exp_baseline = fsub(clim_exp, clim_baseline)
    ens_diff[experiment+'_'+period] = diff_exp_baseline
```

```python
pp_colorbar = dict(proj   = 'Robinson',
                   #colors = '-100 -75 -50 -25 -10 -5 -1 1 5 10 25 50 75 100',
                   colors = '-100 -50 -20 -15 -10 -6 -3 -1 1 3 6 10 15 20 50 100',
                   color  = 'temp_diff_18lev',#NCV_blu_red',
                   mpCenterLonF = 0,
                   #contours = 1,
                   focus = 'land',
                   tiMainFontHeightF=0.03,
                   gsnStringFontHeightF=0.02,
                   gsnRightString='', gsnLeftString=''
                  )
```

```python
pp = pp_colorbar.copy()
pp.update(
    dict(options='gsnAddCyclic=True|lbLabelBarOn=False|mpGridAndLimbOn=True|mpGridLineColor=-1')
)         
```

```python
# -- Plot the individual models
wfuture = 'ssp585_2081-2100'
iplot_members(ens_diff[wfuture], N=1, **pp)
```

```python
# -- Colorbar
def extract_labelbar(figure_file,labelbar_file) :
    #import PIL
    im = PILImage.open(figure_file)
    #box=(left, upper, right, lower).
    im_crop = im.crop((55, 564, 930, 642))
    im_crop.save(labelbar_file, quality=95)

# -- Base plot for the colorbar
wfuture = 'ssp585_2081-2100'
plot_4colorbar = plot(ens_diff[wfuture][ens_diff[wfuture].keys()[0]],
                      options='lbOrientation=horizontal|lbLabelFontHeightF=0.015|lbBoxEndCapStyle=TriangleBothEnds',
                      **pp_colorbar)

# -- color bar file
colorbar_file = '/home/jservon/Chapter12_IPCC/figs/global_figure_12.4/'+variable.upper()+wthres+'_colorbar.png'

# -- Extract the colorbar
extract_labelbar(cfile(plot_4colorbar),colorbar_file)

Image(colorbar_file)
```

## Ensemble statistics

```python
for perc_agreement in ['90','80','66']:

    signif90_dict = dict()
    for exp in exp_list:
        # -- Experiment and period
        experiment = exp['experiment']
        period = exp['period']
        wfuture = experiment+'_'+period

        # -- Mask of models with
        ens_mask_pos = ccdo(ens_diff[wfuture], operator='gtc,0')
        ens_mask_neg = ccdo(ens_diff[wfuture], operator='ltc,0')

        perc_ens_pos = fmul( fdiv( ccdo_ens(regridn(ens_mask_pos, cdogrid='r360x180', option='remapnn'), operator='enssum'), len(ens_diff[wfuture]) ), 100 )
        perc_ens_neg = fmul( fdiv( ccdo_ens(regridn(ens_mask_neg, cdogrid='r360x180', option='remapnn'), operator='enssum'), len(ens_diff[wfuture]) ), 100 )

        # -- Signif90
        signif90_pos = ccdo(perc_ens_pos, operator='gtc,'+perc_agreement)
        signif90_neg = ccdo(perc_ens_neg, operator='gtc,'+perc_agreement)
        signif90_dict[wfuture] = fmul( fadd(signif90_pos, signif90_neg), -1)


    outdir = '/home/jservon/Chapter12_IPCC/figs/global_figure_12.4/'

    pp['tiMainFontHeightF'] = 0.032
    pp['gsnStringFontHeightF'] = 0.03
    #pp['colors'] = '-100 -75 -50 -25 -10 -5 -1 1 5 10 25 50 75 100',
    #pp['colors'] = '-100 -50 -20 -15 -10 -5 -1 -0.5 0.5 1 5 10 15 20 50 100'
    #pp['colors'] = '-100 -50 -20 -15 -10 -6 -3 -1 1 3 6 10 15 20 50 100'
    wfuture = 'ssp585_2041-2060'
    ensmedian_field = ccdo_ens(regridn(ens_diff[wfuture],cdogrid='r360x180'), operator='enspctl,50')
    p1 = plot(ensmedian_field,
              signif90_dict[wfuture],
              shade_above = -0.99999,
              shading_options='gsnShadeHigh=3|gsnShadeFillScaleF=1.0',
              gsnCenterString = 'Change in # days WBGT>'+wthres+'~F34~0~F~C',
              #gsnLeftString = '(d)',
              title = '2041-2060, SSP5-8.5',
              #title = '(d) Change in # days WBGT>33~F34~0~F~C 2041-2060, SSP5-8.5',
              **pp
             )
    cfile(p1, target = outdir + variable.upper()+wthres+'_'+wfuture+'_'+perc_agreement+'perc-agreement.png')

    wfuture = 'ssp585_2081-2100'
    ensmedian_field = ccdo_ens(regridn(ens_diff[wfuture],cdogrid='r360x180'), operator='enspctl,50')
    p2 = plot(ensmedian_field,
              signif90_dict[wfuture],
              shade_above = -0.99999,
              shading_options='gsnShadeHigh=3|gsnShadeFillScaleF=1.0',
              gsnCenterString = 'Change in # days WBGT>'+wthres+'~F34~0~F~C',
              #gsnLeftString = '(e)',
              #title = '(e) 2081-2100, SSP5-8.5',
              title = '2081-2100, SSP5-8.5',
              **pp
             )
    cfile(p2, target = outdir + variable.upper()+wthres+'_'+wfuture+'_'+perc_agreement+'perc-agreement.png')

    wfuture = 'ssp126_2081-2100'
    ensmedian_field = ccdo_ens(regridn(ens_diff[wfuture],cdogrid='r360x180'), operator='enspctl,50')
    p3 = plot(ensmedian_field,
              signif90_dict[wfuture],
              shade_above = -0.99999,
              shading_options='gsnShadeHigh=3|gsnShadeFillScaleF=1.0|gsnAddCyclic=True',
              gsnCenterString = 'Change in # days WBGT>'+wthres+'~F34~0~F~C',
              #gsnLeftString = '(f)',
              #title = '(f) 2081-2100, SSP1-2.6',
              title = '2081-2100, SSP1-2.6',
              **pp
             )
    cfile(p3, target = outdir + variable.upper()+wthres+'_'+wfuture+'_'+perc_agreement+'perc-agreement.png')

    mp = cpage(fig_lines=[[p1,p2,p3]],
                insert=colorbar_file,
                insert_width=400, page_width=1200
              )
    iplot(mp)
```
