## Get the regions using regionmask

```python
import xarray as xr
import numpy as np
import regionmask
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
# -- Either keep data
# -- Or create mask
def extract_region(ds, region_name, return_mask=False):

    dat = ds.Q100
    dat.values = np.array(dat.values, dtype=np.float32)


    #CORDEX_mask_3D = ar6_land.mask_3D(dat)

    # -- Get the AR6 regions
    ar6_land = regionmask.defined_regions.ar6.land

    # -- Get land/sea mask (generic)
    land_110 = regionmask.defined_regions.natural_earth.land_110

    # -- Mask the data
    mask_3D = ar6_land.mask_3D(dat) # AR6 mask
    land_mask = land_110.mask_3D(dat) # Land sea mask
    mask_lsm = mask_3D * land_mask.squeeze(drop=True) # Combine the two

    # -- Mask of the region
    region_mask = mask_lsm.sel(region=list(ar6_land.abbrevs).index(region_name))
    if return_mask:
        return region_mask
    else:
        return dat.where(region_mask)

#r2 = CMIP_mask_3D.isel(region=list(CMIP_mask_3D.abbrevs).index(region_name))

#dat_region = dat.where(region_mask)

def quick_plot(dat):
    proj = ccrs.Robinson()
    ax = plt.subplot(111, projection=proj)

    #tas_CNA.isel(time=0).plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree())
    dat.plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(),add_colorbar=False,cmap='jet')
    ar6_land = regionmask.defined_regions.ar6.land

    text_kws = dict(color="#67000d", fontsize=2, bbox=dict(pad=0.2, color="w"))
    ar6_land.plot(
        ax=ax,
        add_ocean=False,
        line_kws=dict(linewidth=1),
        coastlines=False,
        text_kws=text_kws,
    )

    ax.set_title(region_name)
    ax.coastlines();
def add_lonlat_variable_attributes(filename, variable):
    import os
    cmd = 'ncrename -v .__xarray_dataarray_variable__,'+variable+' -v .region,'+variable+' '+filename+' ; ncatted -O -a coordinates,'+variable+',o,c,"lon lat" -a units,lon,o,c,degrees_east -a units,lat,o,c,degrees_north -a standard_name,lon,o,c,longitude -a standard_name,lat,o,c,latitude '+filename
    print cmd
    os.system(cmd)
def maskout_values(filename):
    cmd = 'cdo addc,1 -setctomiss,1 '+filename+' tmp.nc ; mv tmp.nc '+filename
    print cmd
    os.system(cmd)
```
