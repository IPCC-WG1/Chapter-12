```r
region = 'AFRICA'

res = list()

for (scenario in c('RCP85','RCP45')){
    for (horizon in c('2050','2100')){
        json_file = paste('/home/jservon/Chapter12_IPCC/data/coastal_recession/globalErosionProjections_by_AR6_region_',scenario,'_',horizon,'.json',sep='')
        if (horizon=='2050'){term='mid'}
        if (horizon=='2100'){term='late'}
        wname = paste(scenario,term,sep='_')
        res[[wname]] = list()
        
        json_data <- fromJSON(paste(readLines(json_file), collapse=""))
        for (subregion in names(json_data)){
            if ( json_data[[subregion]]$region==region ){
                res[[wname]][[subregion]] = c(json_data[[subregion]]$q5, json_data[[subregion]]$median, json_data[[subregion]]$q95)
            }#end if
        }#end subregion
    }#end horizon
}#and scenario
```

```r
#res = list()
#res[['RCP85_mid']] = c(-100,-50,100)
#res[['RCP85_late']] = c(-300,-200,-100)
#res[['RCP45_mid']] = c(-70,-20,10)
#res[['RCP45_late']] = c(-200,-100,0)
title = '(a) Some region'

"coastal_plot"=function(res, title){

    # RCP85-RCP45
    general_cex = 1.2
    pcex=1.6
    lwdp = 4
    lwd = 6

    color='dodgerblue2'
    
    par(cex=general_cex)
    par(mar=c(3,4,2,0))
    ylim=c(-1000,600)
    xval = 1:5
    plot(xval, rep(0,length(xval)), type='l', lty=2, lwd=2, ylim=ylim, xlab='', ylab='', xaxt='n', yaxt='n', xaxs='i')
    
    # -- X axis
    axis(1, at=c(2,4), labels=c('mid-term','long-term'), font=2)
    
    # -- Y axis
    axis(2, at=seq(-1000,600,400), font=2)
    
    # -- Title
    mtext(title,3, adj=0, font=2, cex=1.8, line=0.5)
    mtext("Meters",2, font=2, cex=1.3, line=2.5)


    xmid = 2
    xlong = 4
    xadj = 0.1

    resname = 'RCP85_mid'
    print(paste(title,resname))
    print(res[[resname]])
    lines(c(xmid-xadj, xmid-xadj), c(res[[resname]][1], res[[resname]][3]), type='l', lwd=lwd, col=color)
    points(xmid-xadj, res[[resname]][2], pch=16, col=color, cex=pcex)

    resname = 'RCP45_mid'
    print(paste(title,resname))
    print(res[[resname]])
    lines(c(xmid+xadj, xmid+xadj), c(res[[resname]][1], res[[resname]][3]), type='l', lwd=lwd, col=color)
    points(xmid+xadj, res[[resname]][2], pch=16, col='white', cex=pcex)
    points(xmid+xadj, res[[resname]][2], col=color, lwd=lwdp, cex=pcex)

    resname = 'RCP85_late'
    print(paste(title,resname))
    print(res[[resname]])
    lines(c(xlong-xadj, xlong-xadj), c(res[[resname]][1], res[[resname]][3]), type='l', lwd=lwd, col=color)
    points(xlong-xadj, res[[resname]][2], pch=16, col=color, cex=pcex)

    resname = 'RCP45_late'
    print(paste(title,resname))
    print(res[[resname]])
    lines(c(xlong+xadj, xlong+xadj), c(res[[resname]][1], res[[resname]][3]), type='l', lwd=lwd, col=color)
    points(xlong+xadj, res[[resname]][2], pch=16, col='white', cex=pcex)
    points(xlong+xadj, res[[resname]][2], col=color, lwd=lwdp, cex=pcex)

}


"legend_coastal_plot"=function(){

    # RCP85-RCP45
    general_cex = 1.2
    pcex=1.6
    lwdp = 4
    lwd = 6

    color='dodgerblue2'
    
    par(cex=general_cex)
    par(mar=c(3,4,2,0))
    ylim=c(-1000,600)
    xval = 1:5
    plot(xval, rep(0,length(xval)), type='l', col='white',
         lty=2, lwd=2, ylim=ylim, xlab='', ylab='', xaxt='n', yaxt='n', xaxs='i')
    
    # -- 
    
    # -- X axis
    #axis(1, at=c(2,4), labels=c('mid-term','long-term'), font=2)
    
    # -- Y axis
    #axis(2, at=seq(-1000,600,400), font=2)
    
    # -- Title
    mtext('Legend',3, adj=0, font=2, cex=1.8, line=0.5)
    #mtext("Meters",2, font=2, cex=1.3, line=2.5)

    # -- p5/median/p95
    ymedian = 300
    yhalflen = 200
    xmedian = 1.5
    xtextadj = 0.2
    lines(c(xmedian, xmedian), c(ymedian-yhalflen,ymedian+yhalflen), type='l', lwd=lwd, col='black')
    points(xmedian, ymedian, pch=16, col='black', cex=pcex, lwd=lwdp)
    text(xmedian+xtextadj, ymedian+yhalflen, "95th percentile",adj=0)
    text(xmedian+xtextadj, ymedian, "Multi-model median",adj=0)
    text(xmedian+xtextadj, ymedian-yhalflen, "5th percentile",adj=0)
    
    
    # -- CMIP5 - RCP85
    xrcp85 = 1.5
    lenline = 0.8
    yrcp85 = -300
    lines(c(xrcp85, xrcp85+lenline), c(yrcp85,yrcp85), type='l', lwd=lwdp, col=color)
    points(xrcp85+(lenline/2), yrcp85, pch=16, col=color, lwd=lwdp, cex=pcex)
    text(xrcp85+lenline*1.2, yrcp85, "CMIP5 RCP8.5",adj=0, font=2, cex=1.2)
    
    # -- CMIP5 - RCP45
    xrcp45 = 1.5
    yrcp45 = -600
    lines(c(xrcp45, xrcp45+lenline), c(yrcp45,yrcp45), type='l', lwd=lwd, col=color)
    points(xrcp45+(lenline/2), yrcp45, pch=16, col='white', lwd=lwdp, cex=pcex)
    points(xrcp45+(lenline/2), yrcp45, col=color, lwd=lwdp, cex=pcex)
    text(xrcp45+lenline*1.2, yrcp45, "CMIP5 RCP4.5",adj=0, font=2, cex=1.2)
    

}
#legend_coastal_plot()
```

```r
region = 'AFRICA'

res = list()

for (scenario in c('RCP85','RCP45')){
    for (horizon in c('2050','2100')){
        json_file = paste('/home/jservon/Chapter12_IPCC/data/coastal_recession/globalErosionProjections_by_AR6_region_',scenario,'_',horizon,'.json',sep='')
        if (horizon=='2050'){term='mid'}
        if (horizon=='2100'){term='late'}
        wname = paste(scenario,term,sep='_')
        res[[wname]] = list()
        
        json_data <- fromJSON(paste(readLines(json_file), collapse=""))
        for (subregion in names(json_data)){
            if ( json_data[[subregion]]$region==region ){
                res[[wname]][[subregion]] = c(json_data[[subregion]]$q5, json_data[[subregion]]$median, json_data[[subregion]]$q95)
            }#end if
        }#end subregion
    }#end horizon
}#and scenario


figname = paste(region, '_satellite_coastal.png',sep='')
png(figname, width=900,height=1200)
layout(t(matrix(1:12,3,4)))
for (subregion in names(res[[1]])){
    wres = list()
    for (wname in c('RCP85_mid', 'RCP85_late','RCP45_mid','RCP45_late')){
        wres[[wname]] = res[[wname]][[subregion]]
    }
    #wres[['RCP85_mid']] = res[[c(-100,-50,100)
    #wres[['RCP85_late']] = c(-300,-200,-100)
    #wres[['RCP45_mid']] = c(-70,-20,10)
    #wres[['RCP45_late']] = c(-200,-100,0)

    coastal_plot(wres, subregion)
}
legend_coastal_plot()
dev.off()
display_png(file=figname) 
```
