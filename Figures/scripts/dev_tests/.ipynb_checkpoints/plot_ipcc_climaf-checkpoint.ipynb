```python
myplot = plot_ipcc(dat, proj='Robinson',
              color='BlueYellowRed',
              reverse=True,
             options='mpGridAndLimbOn=True|mpGridLineColor=-1')
#  res@mpGridAndLimbOn      = True           ; turn on limb and grid lines
#  res@mpGridLineColor      = -1             ; ...but don't draw grid lines...

iplot(myplot)
```
