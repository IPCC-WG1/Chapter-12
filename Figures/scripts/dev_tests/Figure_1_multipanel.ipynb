```python
import os
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
fig1 = '/home/jservon/test.png'
colorbar = '/home/jservon/colorbar_temp_penalty_perc_diffhorizon2020.png'
mymap = '/home/jservon/colorbar_temp_penalty.png'
```

```python
img2 = mpimg.imread(mymap)
img3 = mpimg.imread(colorbar)

plt.figure(1, figsize=[20,35])#,dpi=500)
mapheight = 4
colorbarheight = 1.5

grid = plt.GridSpec(10, 12, wspace=0.1, hspace=0, height_ratios=[mapheight,colorbarheight,
                                                                 mapheight,colorbarheight,
                                                                 mapheight,colorbarheight,
                                                                 mapheight,
                                                                 mapheight,
                                                                 mapheight,colorbarheight
                                                                ])

# Premiere ligne
wline = 1
plt.subplot(grid[wline*2-2,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,8:12]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-1,4:8]).axis('off')
plt.imshow(img3)

wline = 2
plt.subplot(grid[wline*2-2,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,8:12]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-1,4:8]).axis('off')
plt.imshow(img3)


wline = 3
plt.subplot(grid[wline*2-2,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-2,8:12]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline*2-1,4:8]).axis('off')
plt.imshow(img3)

wline = 6
plt.subplot(grid[wline,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,8:12]).axis('off')
plt.imshow(img2)

wline = 7
plt.subplot(grid[wline,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,8:12]).axis('off')
plt.imshow(img2)

wline = 8
plt.subplot(grid[wline,0:4]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,4:8]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline,8:12]).axis('off')
plt.imshow(img2)

plt.subplot(grid[wline+1,4:8]).axis('off')
plt.imshow(img3)


plt.show()

#plt.savefig("test.png", transparent=True, bbox_inches='tight', pad_inches=0)
```
